```python
import pandas as pd
from wc_simd.dedupe_service import dedup_data_file
dedup_data = pd.read_csv("../" + dedup_data_file, index_col=0)

# force all columns to string type
dedup_data = dedup_data.astype(str)

dedup_data
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("Qwen/Qwen3-Embedding-0.6B")
```

```python
# embed_prompt = "Given a web search query, retrieve relevant passages that answer the query"
```

```python
# embed_prompt = "Given a search query with a person's name, retrieve relevant passages that has the person mentioned."
```

```python
embed_prompt = "INSTRUCT: Given a search query with a person's name, retrieve relevant passages that has the person mentioned. QUERY: "
```

```python
ng_rows = dedup_data[dedup_data['label'].str.contains(
    "florence", case=False, na=False) & dedup_data['label'].str.contains(
    "night", case=False, na=False)]
ng_rows
```

```python
# ng_docs = ["Florence Nightingale", "NIGHTINGALE, FLORENCE"]
# ng_docs = ng_rows["label"].str.lower()
ng_docs = ng_rows["label"]

ng_embeds = model.encode(ng_docs, prompt=embed_prompt)

ng_embeds
```

```python
pd.DataFrame({"label": ng_docs, "embedding": list(ng_embeds)})
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarities = cosine_similarity(ng_embeds)
```

```python
# Create a DataFrame that shows similarity between 2 labels
# similarities = model.similarity(ng_embeds, ng_embeds)
similarity_df = pd.DataFrame(
    similarities,
    columns=ng_rows['label'].values,
    index=ng_rows['label'].values)
similarity_df
```

```python
filtered_similarity_df = similarity_df[similarity_df > 0.75]
filtered_similarity_df
```

```python
# Random sample pairs from dedup_data filtered for type = Person

random_pairs = dedup_data[dedup_data['type'] == 'Person'].sample(n=10)

# Add "Florence Nightingale" to the random pairs dataframe

new_row = pd.DataFrame({'label': ['Florence Nightingale'], 'type': ['Person']})
random_pairs = pd.concat([random_pairs, new_row], ignore_index=True)

# and get embeddings (using hf.embed_documents) and similarities

embeddings = model.encode(random_pairs['label'], prompt=embed_prompt)

similarities = model.similarity(embeddings, embeddings)

# print
print(similarities)

# print number of records > 0.6
print((similarities > 0.63).sum())
```

```python
# Create a DataFrame that shows similarity between 2 labels
similarity_df = pd.DataFrame(similarities, columns=random_pairs['label'].values, index=random_pairs['label'].values)
similarity_df
```

```python
# Filter similarity_df for values greater than 0.6
filtered_similarity_df = similarity_df[similarity_df > 0.65]
filtered_similarity_df
```
