```python
from transformers import AutoModel

gme = AutoModel.from_pretrained(
    "Alibaba-NLP/gme-Qwen2-VL-2B-Instruct",
    trust_remote_code=True
)
```

```python
t2i_prompt = 'Find an image that matches the given text.'

X = gme.get_text_embeddings(texts=["Potted plant"], instruction=t2i_prompt).tolist()
X
```

```python
t2i_prompt = 'Find an image that matches the given text.'

X = gme.get_text_embeddings(texts=["flowers"], instruction=t2i_prompt).tolist()
X
```

```python
# X to numPy
import numpy as np
X_np = np.array(X)
```

```python
from wc_simd.vlm_embed_vae import VAE3DWrapper

sample = False
model_checkpoint = '../../../../runs/vlm_embed_vae3d_hires_1/vae3d.pt'
wrapper = VAE3DWrapper(model_checkpoint)
Z = wrapper.to3d(
    X_np, use_mu=(not sample),
    batch_size=64 * 1024)

Z
```
